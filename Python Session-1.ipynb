# Variable

We don't need to declare a variable before using it. In Python, we simply assign a value to a variable and it will exist. We don't even have to declare the type of the variable. This is handled internally according to the type of value we assign to the variable

Rules for Writing Variables:

1)Variables can be a combination of letters in lowercase (a to z) or uppercase (A to Z) or digits (0 to 9) or an underscore (_).

2)A variable cannot start with a digit. 1variable is invalid, but variable1 is perfectly fine.

3)Keywords cannot be used as variables.

```python
a = 30
b = 1.5
c = "Hello!"
d = True
e = None

print(a,b,c,d,e)
```

# Lists

**Ordered**: Yes<br>
**Mutable**: Yes

List is an ordered sequence of items of any data types.

Declaring a list is , Items separated by commas are enclosed within brackets [ ].

```python
a = [1, 2.5, '3',True]
a[1]=3.5
print(a[1])  
print(len(a))
```

```python
#Slicing
print(a[::-1])
```

```python
# Add to List
a * 2                # [1, 3.5, '3', True, 1, 2, '3', True]
a + [100]            # [1, 3.5, '3', True, 100] --> doesn't mutate original list, creates new one
a.append(100)        # None --> Mutates original list to [1, 3.5, '3', True, 100]          # Or: <list> += [<el>]
a.extend([100, 200]) # None --> Mutates original list to [1, 3.5, '3', True, 100, 200]
a.insert(2, '!!!')   # None -->  [1, 3.5, '!!!', '3', True] - Inserts item at index and moves the rest to the right.

# ' '.join(['Hello','There'])# 'Hello There' --> Joins elements using string as separator.
```

```python
# Copy a List
basket = ['apples', 'pears', 'oranges']
new_basket = basket.copy()
new_basket2 = basket[:]
```

```python
# Remove from List
[1,2,3].pop()    # 3 --> mutates original list, default index in the pop method is -1 (the last item)
[1,2,3].pop(1)   # 2 --> mutates original list
[1,2,3].remove(2)# None --> [1,3] Removes first occurrence of item or raises ValueError.
[1,2,3].clear()  # None --> mutates original list and removes all items: []
del [1,2,3][0]   # None --> removes item on index 0 or raises IndexError
```

```python
# Ordering
[1,2,5,3].sort()         # None --> Mutates list to [1, 2, 3, 5]
[1,2,5,3].sort(reverse=True) # None --> Mutates list to [5, 3, 2, 1]
[1,2,5,3].reverse()      # None --> Mutates list to [3, 5, 2, 1]
sorted([1,2,5,3])        # [1, 2, 3, 5] --> new list created
list(reversed([1,2,5,3]))# [3, 5, 2, 1] --> reversed() returns an iterator
```

```python
# List Comprehensions
# new_list[<action> for <item> in <iterator> if <some condition>]
a = [i for i in 'hello']                  # ['h', 'e', 'l', 'l', '0']
b = [i*2 for i in [1,2,3]]                # [2, 4, 6]
c = [i for i in range(0,10) if i % 2 == 0]# [0, 2, 4, 6, 8]
```

# Sets

**Ordered**: No<br>
**Mutable**: Yes<br>
A set is an unordered collection of items. Every element is unique (no duplicates).

Sets can be used to perform mathematical set operations like union, intersection, symmetric difference etc.

```python
#methods
my_set = set()
my_set.add(1)  # {1}
my_set.add(100)# {1, 100}
my_set.add(100)# {1, 100} --> no duplicates!
my_set.remove(100)      # {1} --> Raises KeyError if element not found
my_set.discard(100)     # {1} --> Doesn't raise an error if element not found
my_set.clear()          # {}
new_set = {1,2,3}.copy()# {1,2,3}
```

```python
#Python Set Operations
set1 = {1,2,3}
set2 = {3,4,5}
set3 = set1.union(set2)               # {1,2,3,4,5}
set4 = set1.intersection(set2)        # {3}
set5 = set1.difference(set2)          # {1, 2}
set6 = set1.symmetric_difference(set2)# {1, 2, 4, 5}
set1.issubset(set2)                   # False
set1.issuperset(set2)                 # False
set1.isdisjoint(set2)                 # False --> return True if two sets have a null intersection.
```

# Dictionary

**Ordered**: No<br>
**Mutable**: Yes<br>

->Python dictionary is an unordered collection of items. While other compound data types have only value as an element, a dictionary has a key: value pair.

->We use curly brackets to contain a dictionary, and colons to indicate keys and values.

```python
my_dict = {'name': 'Sarada Prasad', 'age': 30, 'magic_power': False}
my_dict['name']                      # Sarada Prasad
len(my_dict)                         # 3
list(my_dict.keys())                 # ['name', 'age', 'magic_power']
list(my_dict.values())               # ['Sarada Prasad', 30, False]
list(my_dict.items())                # [('name', 'Sarada Prasad'), ('age', 30), ('magic_power', False)]
my_dict['favourite_snack'] = 'Grapes'# {'name': 'Sarada Prasad', 'age': 30, 'magic_power': False, 'favourite_snack': 'Grapes'}
my_dict.get('age')                   # 30 --> Returns None if key does not exist.
my_dict.get('ages', 0 )              # 0 --> Returns default (2nd param) if key is not found
```

```python
#Remove key
del my_dict['name']
my_dict.pop('name', None)
```

  for element in sequence :

# Flow Chart

```python
my_list = [1,2,3]
my_tuple = (1,2,3)
my_dict = {'a': 1, 'b': 2, 'c': 3}

for num in my_list:
    print(num) # 1, 2, 3

for num in my_tuple:
    print(num) # 1, 2, 3

for num in '123':
    print(num) # 1, 2, 3

for k,v in my_dict.items(): # Dictionary Unpacking
    print(k) # 'a', 'b', 'c'
    print(v) # 1, 2, 3
```

# While Loop

The while loop in Python is used to iterate over a block of code as long as the test expression (condition) is true.

    while test_expression:
        
        Body of while

The body of the loop is entered only if the test_expression evaluates to True.

After one iteration, the test expression is checked again.

This process continues until the test_expression evaluates to False.
